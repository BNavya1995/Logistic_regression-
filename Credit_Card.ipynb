```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split,StratifiedKFold,cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import *
from sklearn.pipeline import Pipeline

```

```python
df=pd.read_csv("/content/creditcard_fraud.csv")
df
```

```python
df.describe()
```

```python
print("Target Distribtuion")
print(df["Class"].value_counts())
```

```python
fraud_count=df['Class'].sum()
legit_count=len(df)-fraud_count
print(f"Legitimate: {legit_count:,} ({legit_count/len(df)*100:.2f}%)")
print(f"Fraud:      {fraud_count:,} ({fraud_count/len(df)*100:.2f}%)")
print(f"\nImbalance ratio: {legit_count//fraud_count}:1")
```

```python
df.isnull().sum().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
x=df.drop("Class",axis=1)
y=df['Class']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
print(f"Train set: {x_train.shape[0]:,} samples")
print(f"Test set:  {x_test.shape[0]:,} samples")
print(f"\nTrain — Fraud: {y_train.sum()}, Legit: {(y_train==0).sum()}")
print(f"Test  — Fraud: {y_test.sum()}, Legit: {(y_test==0).sum()}")
```

```python
sl=StandardScaler()
x_train_scale=sl.fit_transform(x_train)
x_test_scale=sl.transform(x_test)
```

```python

print("Scaling done!")
print(f"  fit_transform on TRAIN only (learned mean/std from train)")
print(f"  transform on TEST (using train's mean/std — no leakage!)")
```

```python
lr_native=LogisticRegression(max_iter=1000,random_state=42)
lr_native.fit(x_train_scale,y_train)

y_pred_native=lr_native.predict(x_test_scale)
y_prob_native=lr_native.predict(x_test_scale)
print("Model trained! Let's check the results...")
```

```python
acc=accuracy_score(y_pred_native,y_test)
prec=precision_score(y_pred_native,y_test)
rec=recall_score(y_pred_native,y_test)
f1 = f1_score(y_pred_native,y_test)
auc=roc_auc_score(y_pred_native,y_test)
print(f"\n  Accuracy:  {acc:.4f}  ({acc*100:.2f}%)")
print(f"  Precision: {prec:.4f}  ({prec*100:.2f}%)")
print(f"  Recall:    {rec:.4f}  ({rec*100:.2f}%)")
print(f"  F1-Score:  {f1:.4f}")
print(f"  ROC-AUC:   {auc:.4f}")
```

```python
cm=confusion_matrix(y_pred_native,y_test)
tn,tp,fn,fp=cm.ravel()
print("\nConfusion Matrix:")
print(f"                    Predicted")
print(f"                  Legit    Fraud")
print(f"  Actual Legit  [{tn:>6,}   {fp:>5,}]")
print(f"  Actual Fraud  [{fn:>6,}   {tp:>5,}]")
```

```python
print("\nFull Classification Report:")
print(classification_report(y_test, y_pred_native, target_names=['Legit', 'Fraud']))
```
